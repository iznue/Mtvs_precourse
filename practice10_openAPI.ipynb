```python
import pandas as pd

df = pd.read_csv('서울시 공영주차장 안내 정보.csv', encoding='cp949') # csv를 불러올 때 encoding 옵션을 처리하는 것이 좋음
```

```python
df.describe()
```

```python
# 각 구별 공영 주차장의 수
print(df['주소'][0])
print(df['주소'][0].split()[0])
```

```python
ku_dict = {}

for temp in df['주소']:
    ku = temp.split()[0]

    if ku in ku_dict.keys():
        ku_dict[ku] = ku_dict[ku] + 1
    else:
        ku_dict[ku] = 1

print(ku_dict)
```

```python
# dictionary -> dataframe
ku_df = pd.DataFrame.from_dict([ku_dict]) # dictionary data -> pandas
ku_df
```

```python
# index와 column 변경
ku_df = ku_df.T
ku_df
```

```python
ku_df.plot(kind='bar')
plt.show()
```

```python
ku_df = ku_df.sort_values(by=0)
ku_df.plot(kind='bar')
plt.show()
```

```python
import requests
import json
import urllib

# string을 url 형식으로
key = urllib.parse.quote('Uwc5dch84o8WvJLITstjcrb9aL5JZbceIHFtu0yoBjx9Ov072bBmoUynIDeFce7NIpbTdNj2lD5f1KSibK8wpQ==')
url = 'https://apis.data.go.kr/B551011/KorService1/searchFestival1?MobileOS=ETC&MobileApp=Meta&serviceKey='+key+'&eventStartDate=20230507&_type=json'
```

```python
response = requests.get(url)
# 인증서 오류가 발생하는 경우 : requests.get(url, verify=False)로 지정
print(response.content)
print(response.text)
```

```python
# json parser에 response.text 실행 값을 넣음
data = json.loads(response.text)
data['response']['body']['items']['item']
```

```python
# 축제의 이름 및 시작일 종료일 주소 전화번호 데이터를 저장
fdata = data['response']['body']['items']['item']

f_list=[]

for temp in fdata:
    temp_dict = {}
    temp_dict['title'] = temp['title']
    temp_dict['start_date'] = temp['eventstartdate']
    temp_dict['end_date'] = temp['eventenddate']
    temp_dict['address'] = temp['addr1']
    temp_dict['tel'] = temp['tel']
    #print(temp['title'])

    f_list.append(temp_dict)

f_list
```

```python
f_df = pd.DataFrame.from_dict(f_list)
f_df
```
